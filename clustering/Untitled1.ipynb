# K-Means clustering!

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
data, category= make_blobs(n_samples = 200, n_features = 2, centers = 4, cluster_std=1.6, random_state=50)

x = np.array(data[:,0])
y = np.array(data[:,1])

print(x.shape)
print(y.shape)
```

```python
plt.scatter(x, y, c = category, cmap = 'viridis')
plt.xlim(-15,15)
plt.ylim(-15,15)
```

```python
kmeans = KMeans(n_clusters = 4)
kmeans.fit(data)
print(kmeans.cluster_centers_)

y_km = kmeans.fit_predict(data)
```
