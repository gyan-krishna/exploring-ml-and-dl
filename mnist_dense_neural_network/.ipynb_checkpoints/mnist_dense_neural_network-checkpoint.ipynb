```python
from keras.models import Sequential
from keras.layers import Dense, Flatten, Dropout, Activation
from keras.utils import np_utils, plot_model

from keras.datasets import mnist
from numpy import argmax
import pydot
import matplotlib.pyplot as pltb
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
plt.imshow(x_train[0])
print(x_train.shape)
print()
```

```python
#normalising the data set
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
```

```python
#creating the output matrix as a categorical dataset
Y_train = np_utils.to_categorical(y_train, 10)
Y_test  = np_utils.to_categorical(y_test, 10)
```

```python
#creating the model

model = Sequential()
model.add(Flatten(input_shape=(28, 28)))
model.add(Dense(128, activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(10))
model.add(Activation('softmax'))

model.summary()
```

```python
plot_model(model,
          to_file = "model.jpg",
          show_shapes = True,
          show_layer_names = True,
          rankdir = "TB",
          expand_nested = False,
          dpi = 96)
```

```python
model.compile(loss = 'categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.fit(x_train, Y_train, batch_size=32, epochs=10, verbose=1)
```

```python
score = model.evaluate(x_test, Y_test, verbose = 1)
#5  epochs = 725us/step - loss: 0.1464 - accuracy: 0.9808
#10 epochs = 687us/step - loss: 0.0634 - accuracy: 0.9818
#20 epochs = 671us/step - loss: 0.0972 - accuracy: 0.9787
#30 epochs = 668us/step - loss: 0.1243 - accuracy: 0.9805
#40 epochs = 808us/step - loss: 0.1477 - accuracy: 0.9821
```

```python
plt.imshow(x_test[0])
error_count = 0
for i in range(len(x_test)):
    x_test0 = x_test[i].reshape(-1,28,28)
    prediction = model.predict(x_test0)
    predicted_class = argmax(prediction)
    if(predicted_class != y_test[i]):
        error_count+=1
print(error_count)
```
